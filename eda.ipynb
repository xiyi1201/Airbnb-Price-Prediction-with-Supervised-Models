### Subset data

```python
df = dataafter2019.drop(columns = ["id", "name", "host_id", "host_name", "neighbourhood", "latitude", "longitude", "availability_365"])
df.head()
```

```python
df['minimum_nights'].min(),df['minimum_nights'].max()
```

```python
df['number_of_reviews'].min(),df['number_of_reviews'].max()
```

### license: Group license to three categories

```python
df['license'].value_counts()
```

```python
def categorize_license(row):
    if row['license'] == 'No License':
        return 'No License'
    elif row['license'] == 'Exempt':
        return 'Exempt'
    else:
        return 'With License'
```

```python
df["license"] = df.apply(categorize_license, axis=1)
df["license"].value_counts()
```

### rating: Change No rating to 0, drop New

```python
df["rating"].value_counts()
```

```python
df["rating"] = df["rating"].replace({"No rating": 0})
```

```python
df = df[df["rating"] != "New "] 
```

```python
df["rating"] = df["rating"].astype(float)
```

### Replace studio as 0, replace baths no specified as 0

```python
df["bedrooms"] = df["bedrooms"].replace({"Studio": 0}).astype(float)
```

```python
df["baths"] = df["baths"].replace({"Not specified": 0}).astype(float)
```

### minimum_nights: Break minimum nights into less than 7 days, 7-30 days, more than 30 days

```python
def categorize_night(row):
    if row['minimum_nights'] <= 7:
        return 'Less than 7'
    elif row['minimum_nights'] <= 30:
        return '7 to 30'
    else:
        return 'More than 30'
```

```python
df["minimum_nights_bins"] = df.apply(categorize_night, axis=1)
```

```python
df["minimum_nights_bins"].value_counts()
```

### last_review: Calculate days to last reviews (1/5/2024) - Recency

```python
df["review_days"] = (pd.to_datetime('2024-01-05') - df['last_review']).dt.days
df
```
